## Fetch Panel Dataset from Google Finance

```python
import pandas as pd
from pandas_datareader import data
```

```python
companies = ["MSFT", "GOOG", "AAPL", "AABA", "AMZN"]
p = data.DataReader(name = companies, data_source = "yahoo", start = "2010-01-01", end = "2016-12-31")
type(p)
```
