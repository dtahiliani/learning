## Set New Values for a Specific Cell or Row

```python
bond.loc["Dr. No", ["Box Office", "Budget", "Bond Actor Salary"]] = [448800000, 700000, 600000]
```

## Set Multiple Values in a DataFrame

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
```

```python
bond
```

```python
mask = bond["Actor"] == "Sean Connery"
bond.loc[mask, "Actor"] = "Sir Sean Connery"
```

```python
bond
```

## Rename Index Labels or Columns in a DataFrame

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
bond.rename({"Year" : "Release Date", "Box Office" : "Revenue"}, inplace = True)
```

```python
bond.head()
```

```python
bond.rename(index = {"Dr. No" : "Doctor No",
                    "GoldenEye" : "Golden Eye",
                    "The World Is Not Enough" : "Best Bond Movie Ever"},
           inplace = True)
```

```python
bond
```

```python
bond.columns
```

```python
bond.columns = ["Year of Release", "Actor", "Director", "Gross", "Cost", "Salary"]
```

```python
bond.head()
```

## The .nsmallest() and .nlargest() Methods

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
bond.sort_values("Box Office", ascending = False).head(3)
```

```python
bond.nlargest(3, "Box Office")
```

```python
bond.nsmallest(3, "Box Office")
```

```python
bond.nlargest(3, "Budget")
```

```python
bond.nsmallest(3, "Bond Actor Salary")
```

```python
bond["Box Office"].nlargest(3)
```

```python
bond["Year"].nsmallest(2)
```
